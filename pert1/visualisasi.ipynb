```python
import matplotlib.pyplot as plt # Untuk memanggil library-nya

flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12, 11, 4, 8, 7)

plt.pie(
    votes,
    labels = flavors,
    autopct='%1.1f%%',
)
plt.show()
```

```python
import matplotlib.pyplot as plt # Untuk memanggil library-nya

flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12, 11, 4, 8, 7)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')

plt.pie(
    votes,
    labels = flavors,
    autopct='%1.1f%%',
    colors=colors,
)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

countries = ('Indonesia', 'Malaysia', 'Brunei', 'Thailand', 'Timor Leste')
populations = (279319404, 34602648, 455191, 71869367, 1413958)

x_coords = np.arange(len(countries))
plt.bar(x_coords, populations, tick_label=countries)
plt.xticks(rotation=360)
plt.show()
```
